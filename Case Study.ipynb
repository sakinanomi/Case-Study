## Maximum things are sold in which month

```python
months= df.groupby(df['order_date'].dt.to_period('M'))
months.mean()
```

```python
def getMonth(s):
  return s.split("-")[1]

df.head()
```

```python
df['month']=pd.Series(df['order_date'], dtype="string")##converting datetime to string

type(df['month'][0])
```

```python
df['month']= df['month'].apply(lambda x: getMonth(x))#making a month column of the months

```

```python
df.head()
```

```python
m=df.groupby('month')#grouping bu months
counts=m.quantity.count() #seeing how many things are sold sccording to month
```

```python
counts.idxmax()#getting index of highest quantity
```

### Thus we can say maximum amount of things are sold in November
